```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as mplt
import pylab as plt
from matplotlib import rc
from pylab import figure, show, legend, xlabel, ylabel
```

```python
df = pd.read_csv("/home/raulmnt/eval1/GLB.Ts.csv",skiprows=[0,139])
```

```python
df.head(20)
```

```python
df.columns= ['Year','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec','J-D','D-N','DJF','MAM','JJA','SON']
```

```python
x=df[u'Year']
y=df[u'J-D']
```

```python
mplt.plot(x,y,'.',color='black')
mplt.plot(x,y,'k-')
mplt.grid(True)
plt.title('Variabilidad en la temperatura terrestre promedio')
plt.ylabel('Temperature anomaly (C))')
```

```python
plt.show()
```

```python
import numpy as np
import statsmodels.api as sm

x=df[u'Year']
y=df[u'J-D']
lowess = sm.nonparametric.lowess(y, x, frac=0.1)
plt.ylabel('Temperature anomaly (C)')
plt.title('Variabilidad suavizada en la temperatura terrestre promedio')
plt.plot(lowess[:, 0], lowess[:, 1])
plt.show()
```

```python
x=df[u'Year']
y=df[u'J-D']
lowess = sm.nonparametric.lowess(y, x, frac=0.1)
plt.plot(lowess[:, 0], lowess[:, 1],color='red')
mplt.plot(x,y,'.',color='black')
mplt.plot(x,y,'k-')
mplt.grid(True)
plt.ylabel('Temperature anomaly (C)')
plt.title('Variabilidad orginal y suavizada en la temperatura terrestre promedio')
plt.show()
```
