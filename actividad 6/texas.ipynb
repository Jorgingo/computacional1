```python
from scipy.fftpack import fft, fftfreq, fftshift
# number of signal points
N = 2184
# sample spacing
T = 1
x = df['Date Time']
y = df[' Water Level']
yf = fft(y)
xf = fftfreq(N, T)
xf = fftshift(xf)
yplot = fftshift(yf)
import matplotlib.pyplot as plt
plt.plot(xf, 1.0/N * np.abs(yplot))
plt.xlim(0, .1)



plt.grid()
plt.title(u'Mareas en Rockport Texas (Enero 2016 - marzo 2016) ')  # Colocamos el título del gráfico
plt.xlabel('Frecuencia (1/hr)')  # Colocamos la etiqueta en el eje x
plt.ylabel('Amplitud (m)')  # Colocamos la etiqueta en el eje y

plt.show()
```
